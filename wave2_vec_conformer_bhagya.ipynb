```python
import numpy as np
import os

# Define the paths and languages
data_path = "/home/bhagya/Data_Bhagya_internship_wav2vec_xlsr_IIITH_23L/"
languages = ["Assamese", "Bengali", "Bodo", "Dogri", "English", "Gujarati", "Hindi", "Kannada",
             "Kashmiri", "Konkani", "Maithili", "Malayalam", "Manipuri", "Marathi", "Nepali",
             "Odia", "Punjabi", "Sanskrit", "Santali", "Sindhi", "Tamil", "Telugu", "Urdu"]

# Function to extract and save data
def extract_and_save(file_prefix, save_path):
    # Load the combined data
    X_data = np.load(data_path + f"X_{file_prefix}_original_VAD30_wav2vec2_xlsr_small_last16_cms.npy")
    y_data = np.load(data_path + f"y1_{file_prefix}_original_VAD30_wav2vec2_xlsr_small_last16_cms.npy")

    # Create a directory to save the extracted language-specific data
    os.makedirs(save_path, exist_ok=True)

    # Extract and save data for each language
    for idx, language in enumerate(languages):
        # Extract data for the current language
        X_data_lang = X_data[y_data == idx]
        y_data_lang = y_data[y_data == idx]

        # Save the extracted data
        np.save(save_path + f"X_{file_prefix}_original_VAD30_wav2vec2_xlsr_small_last16_cms_{language}.npy", X_data_lang.astype(np.float16))
        np.save(save_path + f"y_{file_prefix}_original_VAD30_wav2vec2_xlsr_small_last16_cms_{language}.npy", y_data_lang.astype(np.int16))

        print(f"Extracted and saved data for {language} ({file_prefix})")

    # Verification step to ensure all files are correctly saved
    for language in languages:
        x = np.load(save_path + f"X_{file_prefix}_original_VAD30_wav2vec2_xlsr_small_last16_cms_{language}.npy")
        y = np.load(save_path + f"y_{file_prefix}_original_VAD30_wav2vec2_xlsr_small_last16_cms_{language}.npy")
        print(f"{language} ({file_prefix}): X shape = {x.shape}, y shape = {y.shape}, unique labels = {np.unique(y)}")

# Paths to save the extracted data
train_save_path = "/home/bhagya/Data_Bhagya_internship_wav2vec_xlsr_IIITH_23L/23lang_wav2_vec_features/"
test_save_path = "/home/bhagya/Data_Bhagya_internship_wav2vec_xlsr_IIITH_23L/23lang_wav2_vec_features/"

# Extract and save training data
extract_and_save("train", train_save_path)

# Extract and save test data
extract_and_save("test", test_save_path)
```
